Choice 1: will prompt the user to enter an integer and then recursively count the number of
digits in that integer.

Example 1: Input: 64 Output: 2

Example 2: Input: 252 Output: 3

Example 3: Input: 6 Output: 1

```python
def num_of_digits(num):
    if num//10 == 0:
        return 1
    return 1 + num_of_digits(num//10)
```

Choice 2: will prompt the user to enter a list of integers and then recursively find the maximum
value in the list (you cannot use the max function).

Example 1: Input: [13] Output: 13

Example 2: Input: [1, 6, -3, 1, 14, 9, 12, 24] Output: 24

Example 3: Input: [] Output: 0

```python
def max_list(lst, i, mx):
    if i < len(lst):
        if mx < lst[i]:
            mx = lst[i]
        max_list(lst, i+1, mx)
    else:
        print(f"max = {mx}")
```

Choice 3.2 (For FSD peeps): Given a 2D matrix with uniform column sizes, recursively
compute the count of columns that are normalized. A normalized column is defined as having a
mean value of 0 and a standard deviation of 1 across all its rows (for that same column), the
values in the matrix should be numerical only.

Example:
Input: Matrix = [ [-1.2649110640673518, 5.123451, 43]
[-0.6324555320336759, 5.13123123, 4334]
[0.0, 6.1543453, 125879]
[0.6324555320336759, 0.1231235709, 123544]
[1.2649110640673518, 9.1543524234, 55676] ]

Output: The number of normalized columns is: 1
Explanation: If you calculate the mean and the standard deviation of the first column, you will
see that they are equal to 0 and 1 respectively.

Choice 4: will terminate the program.

Note: you can add extra parameters to the function for each choice.

```python
import numpy as np
def norml(mat, i, nr, nc, n_norm):
    if (i < nc):
        cl = [mat[j][i] for j in range(nr)]
        cl = np.array(cl)
        if np.mean(cl) < 0.00001 and np.mean(cl) >= -0.00001 and np.std(cl) > 0.00001 and np.std(cl) < 1.00001:
            n_norm = n_norm+1
        norml(mat, i+1, nr, nc, n_norm)
    else:
        print(f"The number of normalized columns is: {n_norm}")
```
